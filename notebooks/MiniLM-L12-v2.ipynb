```python
import re
from tqdm import tqdm
tqdm.pandas()
def cleaning_text(text):
    # Приводим текст к нижнему регистру
    text = text.lower()
    # Удаляем ссылки
    text = re.sub(r"http[s]?://\S+", "", text)
    # Удаляем специальные символы и пунктуацию, кроме пробелов
    text = re.sub(r"[^\w\s]", "", text)
    # Удаляем цифры
    text = re.sub(r"\d+", "", text)
    # Удаляем лишние пробелы
    text = re.sub(r"\s+", " ", text).strip()

    return text
df['description'] = df['description'].progress_apply(cleaning_text)
```

```python
from sentence_transformers import SentenceTransformer
import torch
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
df["FullText"] = df["title"] + " " + df["description"]
description_vectors = model.encode(df["FullText"].to_list(), convert_to_tensor=True)
torch.save(description_vectors, "../models/description_vectors_MiniLM-L12-v2.pt")
```

```python
import torch
from sklearn.metrics.pairwise import cosine_similarity

def search_series(query, top_k=5):
    # Кодируем запрос пользователя
    query_vector = model.encode([query], convert_to_tensor=True)

    # Вычисляем косинусное сходство (значения от -1 до 1)
    similarities = cosine_similarity(query_vector.cpu(), description_vectors.cpu())[0]

    # Получаем топ-K индексов наиболее похожих описаний
    top_indices = similarities.argsort()[-top_k:][::-1]

    # Формируем результат с процентом схожести
    results = df.iloc[top_indices][["title", "description"]].copy()
    results["Similarity (%)"] = (similarities[top_indices] * 100).round(2)  # Перевод в проценты

    return results
```

```python
query = "series about detective and investigation of a murdure during war time"
top_series = search_series(query)
display(top_series)
```
